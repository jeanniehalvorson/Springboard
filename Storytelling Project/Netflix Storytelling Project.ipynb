### Dataset Overview
We delved into the Netflix Daily Top 10 dataset, which chronicles the daily rankings of the top 10 movies and TV shows in the United States from 2020 to March 2022. This dataset includes several key attributes such as:

    * Rank: The position of the title on a given day.
    * Title: The name of the movie or TV show.
    * Type: Classification as a movie or TV show.
    * Netflix Exclusive: Indicator if the title is a Netflix Original.
    * Netflix Release Date: The date the title was released on Netflix.
    * Days in Top 10: The number of days a title has remained in the top 10.
    * Viewership Score: A metric representing the popularity of the title.

## 1. Trend of Top-Ranked Titles Over Time

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert 'As of' to datetime
df['As of'] = pd.to_datetime(df['As of'])

# Filter for the top-ranked title (Rank 1)
top_ranked = df[df['Rank'] == 1]

plt.figure(figsize=(12, 6))
sns.lineplot(x='As of', y='Viewership Score', data=top_ranked, marker='o')
plt.title('Trend of Viewership Score of Top-Ranked Titles Over Time')
plt.xlabel('Date')
plt.ylabel('Viewership Score')
plt.show()
```

## 2. Distribution of Viewership Scores

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Viewership Score'], bins=30, kde=True)
plt.title('Distribution of Viewership Scores')
plt.xlabel('Viewership Score')
plt.ylabel('Frequency')
plt.show()
```

## 3. Days in Top 10 by Type (Movie/TV Show)

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Type', y='Days In Top 10', data=df)
plt.title('Average Number of Days in Top 10 by Type')
plt.xlabel('Type')
plt.ylabel('Average Days in Top 10')
plt.show()
```
